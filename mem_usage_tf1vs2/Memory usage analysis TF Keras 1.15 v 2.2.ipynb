```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
labels = ['TF 1.15', 'TF 2.2']
x = np.arange(len(labels)) 
y = [25.0234375, 357.5546875]

fig, ax = plt.subplots()
ax.bar(x, y)

ax.set_ylabel('SavedModel memory usage (MB)')
ax.set_title('Comparing memory usage after saving SavedModel for TF 1 vs 2')
ax.set_xticks(x)
ax.set_xticklabels(labels)
fig.tight_layout()

plt.show()
print('TF2 vs TF1: {:.2f}'.format(y[1] / y[0]))
```

```python
# on-disk size of saved_model.pb
y2 = [0.144, 2.4]

fig, ax = plt.subplots()
ax.bar(x, y2)

ax.set_ylabel('SavedModel on-disk size (MB)')
ax.set_title('Comparing memory usage of SavedModel on-disk for TF 1 vs 2')
ax.set_xticks(x)
ax.set_xticklabels(labels)
fig.tight_layout()

plt.show()
print('TF2 vs TF1: {:.2f}'.format(y2[1] / y2[0]))
```
